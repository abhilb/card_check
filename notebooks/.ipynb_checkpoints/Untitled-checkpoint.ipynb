# Create card images with different backgrounds and at different positions

```python
card_images = list(CARD_IMGS_PATH.rglob("*.png"))
bkgd_images = list(BKGD_IMGS_PATH.rglob("*.jpg"))
```

```python
def get_random_pos(bkgd_fname, card_fname):
    card = Image.open(card_fname)
    bkgd = Image.open(bkgd_fname)
    assert card.width < bkgd.width, "Width of card is greater than the background"
    assert card.height < bkgd.height, "Height of the card is greater than the background"
    x_lims = 0, bkgd.width - (card.width + 5)
    y_lims = 0, bkgd.height - (card.height + 5)
    card.close()
    bkgd.close()
    return (random.randint(*x_lims), random.randint(*y_lims))
```

```python
def paste_card_img_at_random_pos(bkgd_fname, card_fname):
    dest = get_random_pos(bkgd_fname, card_fname)
    bkgd = Image.open(bkgd_fname)
    bkgd = bkgd.convert("RGBA")
    card = Image.open(card_fname)
    newimg = bkgd.copy()
    newimg.alpha_composite(card, dest=dest)
    return newimg
```

```python
idx = 0
for card in card_images:
    for bkgd in bkgd_images:
        for _ in range(50):
            img = paste_card_img_at_random_pos(bkgd, card)
            with open(TRAIN_PATH / f"{idx:05d}.jpg", "wb") as fp:
                img = img.convert("RGB")
                img.save(fp)
            idx += 1
```
